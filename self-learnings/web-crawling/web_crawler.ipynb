```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
```

```python
START_URL = "https://github.com/Anuradhatiwari01"
MAX_PAGES = 100
OUTPUT_FILE = "titles.txt"
```

```python
frontier = [START_URL]
visited = set()
```

```python
output = open(OUTPUT_FILE, "w", encoding = "utf-8")
print("Output file opened:", OUTPUT_FILE)
```

```python
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Educational crawler for IR course)"
}
```

```python
while frontier and len(visited)< MAX_PAGES:
    url = frontier.pop(0)

    if url in visited:
        continue
    print("crawling: ", url)

    visited.add(url)

    try:
        response = requests.get(url, headers = HEADERS)
    except:
        print("Failed to fetch page")
        continue

    soup = BeautifulSoup(response.text, "html.parser")

    if soup.title:
        title = soup.title.get_text().strip()
    else:
        title = "No title"
    print(" Title: ", title)

    output.write(url + "\n")
    output.write(title + "\n\n")

    links = soup.find_all("a")

    for link in links:
        href = link.get("href")
        if not href:
            continue

        absolute_url = urljoin(url, href)

        frontier.append(absolute_url)

output.close()
print("\nDone!!")
print("Total pages crawled: ", len(visited))

```
